The next step is to get to the primary objective of the project: to be able to predict review ratings based on the text of the reviews. It is important to examine the most common phrases and bigrams of all the review text data. In order to do this, we need to implement a trigram vectorizer. 

```python
from sklearn.feature_extraction.text import CountVectorizer

# Instantiate a trigram vectorizer
cv_trigram_vec = CountVectorizer(max_features=200, 
                                 stop_words='english', 
                                 ngram_range = (1,2))

# Fit and apply trigram vectorizer
cv_trigram = cv_trigram_vec.fit_transform(df['reviews.text'])

# Print the trigram features
print(cv_trigram_vec.get_feature_names())
```

In this step, I'm rearranging and creating new categories from the review ratings. This is because it is hard to distinguish reviews say if they were 1 star apart even for a human. I have decided to categorize 4 or 5 star reviews into 'High', 3 stars into 'Medium', 1 or 2 stars into 'Low'.

```python
df['reviews.label'] = 'High'
df.loc[df['reviews.rating'] == 3, 'reviews.label'] = 'Medium'
df.loc[df['reviews.rating'] == 2, 'reviews.label'] = 'Low'
df.loc[df['reviews.rating'] == 1, 'reviews.label'] = 'Low'
```

```python
df['reviews.label'].value_counts()
```

There is an abundance of 5 star ratings, so I decided to approach this imbalance by downsampling the 'High' reviews to approximately double of the other categories. 

```python
from sklearn.utils import resample


X = pd.concat([df['reviews.text'],df['reviews.label']], axis = 1)

rating3 = X[X['reviews.label'] == 'High']
rating2 = X[X['reviews.label'] == 'Medium']
rating1 = X[X['reviews.label'] == 'Low']
rating3_downsample = resample(rating3,
                          replace=True, 
                          n_samples=2* len(rating2), 
                          random_state=0)

downsampled = pd.concat([rating3_downsample,rating2,rating1])
y = downsampled['reviews.label']
X = downsampled.drop('reviews.label', axis = 1)
X
```

I am preparing the text data for vectorization and TFIDF transformation and then performing a train test split with a split of 20% on the traning data. Then I am going to run through several algorithms of choice and graphing a box plot of the best f1 scores to see which algorithms performed best on f1 score. I decided to go with f1 score because it is a better metric of model evaluation for unbalanced data as our data is still somewhat imbalanced.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.2)
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train['reviews.text'])
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
```

```python
#XGBoost requires encoded labels
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
import xgboost as xgb
plt.figure(figsize = (12,5))
from sklearn.model_selection import cross_val_score
models = [
    RandomForestClassifier(max_depth=3, random_state=0),
    LinearSVC(),
    MultinomialNB(),
    LogisticRegression(random_state=0),
    xgb.XGBClassifier(objective='multi:softmax',num_class = 3, random_state =0, eval_metric = 'mlogloss')
]
CV = 5
cv_df = pd.DataFrame(index=range(CV * len(models)))
entries = []
for model in models:
  model_name = model.__class__.__name__
  f1micros = cross_val_score(model, X_train_tfidf, y_train, scoring='f1_micro', cv=CV)
  for fold_idx, f1micro in enumerate(f1micros):
    entries.append((model_name, fold_idx, f1micro))
cv_df = pd.DataFrame(entries, columns=['model_name', 'fold_idx', 'f1_micro'])
import seaborn as sns
sns.boxplot(x='model_name', y='f1_micro', data=cv_df)
sns.stripplot(x='model_name', y='f1_micro', data=cv_df, 
              size=8, jitter=True, edgecolor="gray", linewidth=2)
plt.show()
```

It seems as if Linear SVC performed the best in micro f1 score. Micro f1 score takes into consideration of unbalanced classes and adds weights to each of the classes. Linear SVC seems to have performed the best but however there is a lot more complexity and learning in XGBoost and it has proved to be the more reliable algorithm. It has mostly been the algorithm of choice for competition winning models on Kaggle. I decided to tune XGBoost parameters to see if I could get a better result than LinearSVC.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(sublinear_tf=True, norm='l2',min_df = 2, encoding='latin-1', ngram_range=(1, 2), stop_words='english', token_pattern=r'\b[^\d\W]+\b')

features = tfidf.fit_transform(X['reviews.text']).toarray()
labels = y
features.shape
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Conducting a random search on the XGBoost Classifier

```python
import time

import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
clf = xgb.XGBClassifier(objective='multi:softmax',num_class = 3, random_state =0, eval_metric = 'mlogloss')


param_grid = {
        'max_depth': [6, 10, 15, 20],
        'learning_rate': [0.001, 0.01, 0.1, 0.2, 0.3],
        'subsample': [0.5, 0.6, 0.7, 0.8, 0.9, 1.0],
        'colsample_bytree': [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0],
        'colsample_bylevel': [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0],
        'min_child_weight': [0.5, 1.0, 3.0, 5.0, 7.0, 10.0],
        'gamma': [0, 0.25, 0.5, 1.0],
        'reg_lambda': [0.1, 1.0, 5.0, 10.0, 50.0, 100.0],
        'n_estimators': [50,100,150]}



rs_clf = RandomizedSearchCV(clf, param_grid, n_iter=20,
                            n_jobs=1, verbose=2, cv=2,
                            scoring='f1_micro', refit=False, random_state=42)
print("Randomized search..")
search_time_start = time.time()
rs_clf.fit(X_train, y_train)
print("Randomized search time:", time.time() - search_time_start)

best_score = rs_clf.best_score_
best_params = rs_clf.best_params_
print("Best score: {}".format(best_score))
print("Best params: ")
for param_name in sorted(best_params.keys()):
    print('%s: %r' % (param_name, best_params[param_name]))
```
